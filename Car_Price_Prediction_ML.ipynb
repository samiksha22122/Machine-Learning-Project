```python
data_no_multicolinearity = data_cleaned.drop(['Year'],axis=1)
```

```python
data_no_multicolinearity
```

```python
# Deal with categorical variables
```

```python
data_no_multicolinearity.describe(include='all')
```

```python
data_with_dummies = pd.get_dummies(data_no_multicolinearity,drop_first=True)
```

```python
data_with_dummies
```

```python
y = data_with_dummies['log_price']
```

```python
y
```

```python
X = data_with_dummies.drop(['log_price'],axis=1)
```

```python
X
```

```python
y
```

```python
#standardization
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
X=scaler.transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train,y_train)
```

```python
# R2
```

```python
lin_reg.score(X_train,y_train)
```

```python
# prediction on test data
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
lin_reg.score(X_test,y_test)
```
